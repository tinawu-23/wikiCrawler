```python
from bs4 import BeautifulSoup
import requests
```

```python
url = input("Enter a link: ")
```

```python
source=requests.get(url)
```

```python
data=source.text
```

```python
soup=BeautifulSoup(data,"html.parser")
```

```python
for link in soup.find_all('a'):
    if (link.get('href','').startswith('/wiki/')):
        print(link.get('href'))
```
